```python
## Master 3: Visualization with Pandas
## Description: Various charting using Pandas and matplotlib
## Language: Python
## Author: Alexander Hepburn
## Date: 23.04.2024

# In this task you will visualize your data
# you should work with jupyter notebook! start it before doing the task.

#Import the necessary libraries

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
# we will use data from Titanic Kaggle data set: https://raw.githubusercontent.com/mslyon/mariorepo/master/train.csv
url = 'https://raw.githubusercontent.com/mslyon/mariorepo/master/train.csv'

titanic = pd.read_csv(url)
titanic.head()
```

```python
# Lets set Set PassengerId as the index
titanic.set_index('PassengerId').head()
```

```python
# now we want to create a pie chart presenting the male/female proportion
# sum the instances of males and females
males = (titanic['Sex'] == 'male').sum()
females = (titanic['Sex'] == 'female').sum()

# put them into a list called proportions
proportions = [males, females]
```

```python
# Create a pie chart
# using proportions 
# with the labels being officer names

plt.pie(proportions, labels = ['Males', 'Females'], 

# with no shadows
shadow = False,
# with colors
colors = ['blue','red'],

# with one slide exploded out
explode = (0.15 , 0),

# with the start angle at 90%
startangle = 90,

# with the percent listed as a fraction
autopct = '%1.1f%%')

# View the plot drop above
plt.axis('equal')

# Set labels
plt.title("Sex Proportion")
# View the plot 
plt.tight_layout()
plt.show()
```

```python
# now we want to create a scatterplot with the Fare payed and the Age, differ the plot color by gender

# the code is
# creates the plot using
lm = sns.lmplot(x = 'Age', y = 'Fare', data = titanic, hue = 'Sex', fit_reg=False)

# set title
lm.set(title = 'Fare x Age')
# get the axes object and tweak it
axes = lm.axes
axes[0,0].set_ylim(-5,)
axes[0,0].set_xlim(-5,85)
```

```python
# TASK 1: How many people survived?

survived_df = (titanic['Survived'] == 1)
print(f"{survived_df.sum()} people survived!")
survived_df.head()
```

```python
# TASK 2: can you modify the previsous pie chart to display only the survived

male_survived_df = titanic.loc[(titanic['Survived'] == 1) & (titanic["Sex"] == "male")] # Filter titanic by survived and sex for males
male_survived_sum = male_survived_df.shape[0] # get the number of rows for the amount 

female_survived_df = titanic.loc[(titanic['Survived'] == 1) & (titanic["Sex"] == "female")] # Filter titanic by survived and sex for females
female_survived_sum = female_survived_df.shape[0] # get the number of rows for the amount 

survived_proportions = [male_survived_sum, female_survived_sum] # Create a list with both sums
print(survived_proportions) # Print the number to the console
```

```python
# Same code as before except with survived_proportions
plt.pie(survived_proportions, labels = ['Males', 'Females'], shadow = False, colors = ['blue','red'], explode = (0.15 , 0),
startangle = 90, autopct = '%1.1f%%')


plt.axis('equal') # View the plot drop above
plt.title("Sex Proportion") # Set labels

# View the plot 
plt.tight_layout()
plt.show()
```
